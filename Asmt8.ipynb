```python
import numpy as np
import random
import matplotlib.pyplot as plt

N = 10000
mu = 10

z = []
for i in range(N+1):
    z.append(random.random())

x = []
for j in z:
    x.append(-(np.log(1-j))/mu)

plt.hist(z, bins=50, edgecolor='black')
plt.xlabel("z")
plt.ylabel("Frequency")
plt.show()

plt.hist(x, bins=50, color='red',density =True, edgecolor='black')
plt.xlabel("x")
plt.ylabel("Frequency")

x_val = np.linspace(0,np.max(x)*1.1, 200)
pdf= mu*np.exp(-mu*x_val)
plt.plot(x_val,pdf,linewidth=2)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

N = 10000
miu = 10
sigma = 5

z1 = []
z2 = []
for i in range(N+1):
    z1.append(random.random())
    z2.append(random.random())


r = []
for j in z1:
    r.append(np.sqrt(-2*(sigma)**2*np.log(1-j)))

theta = []
for k in z2:
    theta.append(2*np.pi*k)
    
ax = r*np.cos(theta)

x = sigma * ax + miu

plt.hist(z1, bins=50, edgecolor = 'black')
plt.xlabel("z1")
plt.ylabel("frequency")
plt.show()

plt.hist(x, bins=50, density=True, edgecolor='black')
plt.xlabel("x")
plt.ylabel("density")


x_val = np.linspace(np.min(x), np.max(x), 500)
pdf = np.exp(-((x_val - miu)**2) / (2 * sigma**2)) / (np.sqrt(2*np.pi*sigma**2))

plt.plot(x_val, pdf, color='red', linewidth=2)
plt.show()
```
