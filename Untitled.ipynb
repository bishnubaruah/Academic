```python
import numpy as np
import matplotlib.pyplot as plt

# Grid size (including boundaries)
nx, ny = 5, 5  # 3 interior points + 2 boundaries

# Initialize temperature array
T = np.zeros((ny, nx))

# Apply boundary conditions
T[0, :] = 100.0        # Top boundary (y = max)
T[-1, :] = 0.0         # Bottom boundary (y = 0)
T[:, 0] = 75.0         # Left boundary (x = 0)
T[:, -1] = 50.0        # Right boundary (x = max)

# Initial guess for interior points
T[1:-1, 1:-1] = 50.0

# Relaxation parameter (1 = Gauss-Seidel, >1 = over-relaxation)
omega = 1.0

# Iteration settings
tolerance = 1e-4
max_iter = 10000

for iteration in range(max_iter):
    T_old = T.copy()
    
    # Update all interior points
    for i in range(1, ny-1):
        for j in range(1, nx-1):
            T[i, j] = (1 - omega) * T[i, j] + omega * 0.25 * (
                T[i+1, j] + T[i-1, j] + T[i, j+1] + T[i, j-1]
            )

    # Check for convergence
    error = np.max(np.abs(T - T_old))
    if error < tolerance:
        print(f"Converged after {iteration} iterations with max error = {error:.5e}")
        break
else:
    print("Did not converge within the maximum number of iterations")

# Flip vertically to match Cartesian orientation (bottom = 0°C)
T_plot = np.flipud(T)

# Display results
print("\nTemperature Distribution (°C):")
print(np.round(T_plot, 2))

# Plot the temperature field
plt.imshow(T_plot, cmap='plasma', origin='lower')
plt.colorbar(label='Temperature (°C)')
plt.title('Steady-State Temperature Distribution (Liebmann’s Method)')
plt.xlabel('x-direction')
plt.ylabel('y-direction')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid and initialization
n = 5  # number of grid points (including boundaries)
Told = np.zeros((n, n))

# Boundary conditions
Told[0, 1:-1] = 100   # top boundary (row 0)
Told[-1, :] = 0       # bottom boundary (last row)
Told[1:-1, 0] = 75    # left boundary (column 0)
Told[1:-1, -1] = 50   # right boundary (last column)

# Over-relaxation factor
lmbda = 1.5

# Copy for new values
Tnew = np.copy(Told)

# Convergence criterion
tol = 1e-4
error = 1
iteration = 0
max_iter = 10000

# Iterative Liebmann’s method (SOR)
while error > tol and iteration < max_iter:
    iteration += 1
    error = 0

    for i in range(1, n - 1):
        for j in range(1, n - 1):
            old_val = Told[i, j]
            new_val = 0.25 * (Told[i + 1, j] + Told[i - 1, j] +
                              Told[i, j + 1] + Told[i, j - 1])
            # Successive over-relaxation update
            Tnew[i, j] = (1 - lmbda) * old_val + lmbda * new_val

            # Track maximum error
            error = max(error, abs(Tnew[i, j] - old_val))

    # Update grid for next iteration
    Told[:, :] = Tnew[:, :]

print(f"Converged in {iteration} iterations with final error = {error:.6f}")

# Display temperature matrix
print("\nFinal Temperature Distribution (°C):")
print(np.round(Tnew, 2))

# Optional: flip vertically for correct top/bottom display
T_plot = np.flipud(Tnew)

# Plot the steady-state heat distribution
plt.figure(figsize=(6, 5))
plt.imshow(T_plot, cmap='hot', interpolation='nearest')
plt.colorbar(label="Temperature (°C)")
plt.title("Steady-State Temperature Distribution (Liebmann’s Method)")
plt.show()
```
