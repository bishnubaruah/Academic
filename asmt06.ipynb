## Question 1

```python
fx_0 = 0.000000
fx_1 = 1.386294
fx_2 = 1.791759
fx_3 = 1.609438
```

```python
# Linear interpolation
x_0 = 1
x_1 = 4
x = 2

b_0 = fx_0
b_1 = (fx_1 - fx_0)/(x_1 - x_0)

fx_l = b_0 +b_1*(x-x_0) 
fx_l
```

```python
def fxl(x):
    return b_0+b_1*(x-1)
```

```python
# Quadratic interpolation
x_2 = 6

m = (fx_2 - fx_1)/(x_2 - x_1)
n = (fx_1 - fx_0)/(x_1 - x_0)

b_2 = (m-n)/(x_2-x_0)

fx_q = b_0+b_1*(x-x_0)+b_2*(x-x_0)*(x-x_1)
fx_q
```

```python
def fxq(x):
    return b_0+b_1*(x-1)+b_2*(x-1)*(x-4)
```

```python
fxq(2)
```

```python
# Third order newton's interpolation
x_3 = 5

p = (fx_3-fx_2)/(x_3-x_2)
q = (fx_2-fx_1)/(x_2-x_1)

b_3 = (((p-q)/(x_3-x_1)) - b_2)/(x_3-x_0)

fx_n = b_0+b_1*(x-x_0)+b_2*(x-x_0)*(x-x_1)+b_3*(x-x_0)*(x-x_1)*(x-x_2)
fx_n
```

```python
def fxn(x):
    return b_0+b_1*(x-1)+b_2*(x-1)*(x-4)+b_3*(x-1)*(x-4)*(x-6)
```

```python
import matplotlib.pyplot as plt
import math, numpy

v = numpy.linspace(1,3,100)
y = []
fxli = []
fxqu = []
fxnt = []
for i in v:
    y.append(math.log(i))
    fxli.append(fxl(i))
    fxqu.append(fxq(i))
    fxnt.append(fxn(i))

ln2 = math.log(2)

plt.figure(figsize=(8, 4), dpi=100) 


plt.scatter(2, ln2, color='k', s=30, label='true')
plt.scatter(2, fx_l, color='g', s=30, label='linear') 
plt.scatter(2, fx_q, color='c', s=30, label='quadratic') 
plt.scatter(2, fx_n, color='m', s=30, label='newton') 

plt.plot(v,y, color='k', label="ln(x)")
plt.plot(v,fxli, label ='linear')
plt.plot(v,fxqu, label='quadratic')
plt.plot(v,fxnt, label='third order')

plt.axhline(y=ln2, color='r', linestyle='--', linewidth=1)
plt.axvline(x=2, color='r', linestyle='--', linewidth=1)
plt.xlabel('X')
#plt.ylabel('ln(x)')
plt.legend()
plt.grid()
```
